```python
import ast
ast.literal_eval()
```

```python
def make(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L   
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action' ,'Adventure', 'Fantasy','Science Fiction']
```

```python
make('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']= movies['genres'].apply(make)
```

```python
def  convert(obj):
    List =[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            List.append(i['name'])
            counter +=1
        else:
            break
        
    return List   
```

```python
 movies['cast'].apply(convert)
```

```python
movies['cast']= movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
def  extract_dir(obj):
    List =[]
   
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            List.append(i['name'])
            break
        
    return List
```

```python
movies['crew'].apply(extract_dir)
```

```python
movies['crew']=movies['crew'].apply(extract_dir)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" " ,"") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" " ,"") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id' , 'title', 'tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']= new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    root_word=[]
    for i in text.split():
        root_word.append(ps.stem(i))
        return " ".join(root_word)
        
```

```python
ps.stem('engaging')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])), reverse=True , key=lambda x:x[1])
```

```python
def suggest(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True , key =lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
  
```

```python
suggest('Prometheus')
```
